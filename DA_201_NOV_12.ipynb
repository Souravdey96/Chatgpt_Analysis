# Cleaning the Text

```python
df['clean_review'] = df['Review']
df['clean_review'] = df['clean_review'].str.lower()
```

```python
# Remove Punctuation
df['clean_review'] = df['clean_review'].str.replace(r'[^\w\s]', '', regex=True)
```

```python
#Remove Extra Spaces
df['clean_review'] = df['clean_review'].str.replace(r'\s+', ' ', regex=True).str.strip()
```

# Text Blob Sentiment Analysis

```python
import nltk
nltk.download('punkt')
```

```python
# importing Textblob
from textblob import TextBlob

#Creating polarity and subjectivity
df['polarity'] = df['clean_review'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['subjectivity'] = df['clean_review'].apply(lambda x: TextBlob(x).sentiment.subjectivity)
```

```python
def get_sentiment(score):
  if score > 0:
    return 'Positive'
  elif score < 0:
    return 'Negative'
  else:
    return 'Neutral'
df['sentiment_category'] = df['polarity'].apply(get_sentiment)
```

```python
# Visualize Sentiment
plt.figure(figsize=(6, 4))
sns.countplot(x='sentiment_category', data=df, palette='Set2')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Category')
plt.ylabel('Count of Reviews')
plt.show()
```

# What Users Love

```python
# Filtering Positive Reviews
positive_reviews = df[df['sentiment_category'] == 'Positive']
```

### Combine all positive clean reviews

```python
all_text = " ".join(positive_reviews['clean_review'].astype(str)).lower()

#remove symbols, keep letters/spaces
import re

cleaned_text = re.sub(r'[^a-zA-Z\s]', '', all_text)
words = cleaned_text.split()

#remove short words
words = [word for word in words if len(word) > 2]

#extract bigrams
phrases = []

for i in range(len(words) - 1):
    if len(words[i]) > 2 and len(words[i+1]) > 2:
        phrases.append(f"{words[i]} {words[i+1]}")


#extract trigrams
for i in range(len(words) - 2):
    if len(words[i]) > 2 and len(words[i+1]) > 2 and len(words[i+2]) > 2:
        phrases.append(f"{words[i]} {words[i+1]} {words[i+2]}")

```

### Visualize top 10 Phrases

```python
#count top 10 phrases
from collections import Counter

phrase_freq = Counter(phrases)
pos_phrases = phrase_freq.most_common(10)
pos_phrases

#Visualize top 10 phrases
import matplotlib.pyplot as plt

phrases, phrase_counts = zip(*pos_phrases)

plt.figure(figsize=(10,6))
plt.barh(range(len(phrases)), phrase_counts, color='lightgreen', alpha=0.7)
plt.yticks(range(len(phrases)), phrases)
plt.title("Most Common Phrases in Positive Reviews", fontsize=14, fontweight='bold')
plt.xlabel("Frequency")
plt.tight_layout()
plt.show()
```

# What Frustates Users

```python
# Filtering Negative Rows
negative_reviews = df[df['sentiment_category'] == 'Negative']
```

### Combine all negative clean text

```python
neg_text = " ".join(negative_reviews['clean_review'].astype(str)).lower()
```

### Cleaning Symbols

```python
cleaned_neg_text = re.sub(r'[^a-zA-Z\s]', '', neg_text)
neg_words = cleaned_neg_text.split()
```

###Remove Short words

```python
neg_words = [word for word in neg_words if len(word) > 2]
```

###Extract Negative Bigrams

```python
neg_phrases = []

for i in range(len(neg_words) - 1):
    if len(neg_words[i]) > 2 and len(neg_words[i+1]) > 2:
        neg_phrases.append(f"{neg_words[i]} {neg_words[i+1]}")
```

### Extract Negative Trigrams

```python
for i in range(len(neg_words) - 2):
  if len(neg_words[i]) > 2 and len(neg_words[i+1]) > 2 and len(neg_words[i+2]) > 2:
    neg_phrases.append(f"{neg_words[i]} {neg_words[i + 1]} {neg_words[i + 2]}")
```

### Count the most common negative phrases

```python
neg_phrase_freq = Counter(neg_phrases)
neg_top_phrases = neg_phrase_freq.most_common(10)
neg_top_phrases
```

### Visualize top 10 phrases

```python
neg_phrases_list, neg_phrase_counts = zip(*neg_top_phrases)

plt.figure(figsize=(10,6))
plt.barh(range(len(neg_phrases_list)), neg_phrase_counts, color='tomato', alpha=0.7)
plt.yticks(range(len(neg_phrases_list)), neg_phrases_list)
plt.title("Most Common Phrases in Negative Reviews", fontsize=14, fontweight='bold')
plt.xlabel("Frequency")
plt.tight_layout()
plt.show()
```
